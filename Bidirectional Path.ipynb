```python
import numpy as np

def print_neighbors_with_distance(matrix, max_distance):
    rows, cols = len(matrix), len(matrix[0])

    for i in range(rows):
        for j in range(cols):
            # Neighbors
            neighbors = []

            # Iterate over all possible neighbors within the max_distance
            for x in range(max(0, i - max_distance), min(i + max_distance + 1, rows)):
                for y in range(max(0, j - max_distance), min(j + max_distance + 1, cols)):
                    if (x, y) != (i, j) and abs(x - i) + abs(y - j) <= max_distance:
                        neighbors.append(matrix[x][y])

            # Print the result for the current element
            #print(f"At ({i},{j}): {matrix[i][j]}, Neighbors: {neighbors}")
            for k in neighbors:
                print(matrix[i][j],k)
            

# Example usage:
matrix = np.arange(1, 65).reshape((8, 8))
max_distance = 1
print_neighbors_with_distance(matrix, max_distance)
print(matrix)
```
